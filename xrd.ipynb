Nous allons déterminer les 3 premiers pics du diffractogramme du KScO2 pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å. Nous allons également déterminer les indices hkl associés à ces 3 pics. Les rayons indiqués précédemment vont créer des interférences destructives ou constructives, selon la direction. Il y a donc un phénomène de diffraction, pour lequel on va considérer un angle de $2\theta$ entre la direction d'incidence et celle d'observation.

```python
from pymatgen.ext.matproj import MPRester

with MPRester("OhzYzXS8KZ2GAhVz3g1zkqH8EJz32Y7X") as m:
    # Récupération de la structure du cristal KScO2
    structure = m.get_structure_by_material_id("mp-8188")
```

```python
# On crée un objet XRDCalculator
from pymatgen.analysis.diffraction.xrd import XRDCalculator
xrd = XRDCalculator(1.54060)

# On calcule le modèle de diffraction
pattern = xrd.get_pattern(structure)
```

```python
#On affiche le diffractogramme
xrd.show_plot(structure, two_theta_range=(0, 175), annotate_peaks=False)
```

```python
# On récupère les infos des 3 premiers pics

for i in range(3):
    print("Données du pic numéro "+ str(i+1))
    print("Angle du pic : "+ str(pattern.x[i]))
    print("Intensité du pic : "+ str(pattern.y[i]))
    print("Indices hkl du pic : "+ str(pattern.hkls[i][0]["hkl"]))
    print("Multiplicité du pic : "+ str(pattern.hkls[i][0]["multiplicity"]))
    print("   ")
```
